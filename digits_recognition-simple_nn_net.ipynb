Define DataLoaders from the DataLoader for training data and the DataLoader from the validation data

```python
digit_dls = DataLoaders(train_dl, valid_dl)
```

Define batch accuracy

```python
def batch_accuracy(xb, yb):
    preds = xb.sigmoid()
    correct = (preds>0.5) == yb
    return correct.float().mean()
```

Define a loss function

```python
def digit_recognition_rmse_loss(predictions, targets):
    predictions = predictions.sigmoid()
    return ((predictions-targets)**2).mean().sqrt()
```

Let's define weights and biases for two layers

```python
simple_net = nn.Sequential(
    nn.Linear(28*28,30),
    nn.ReLU(),
    nn.Linear(30,10)
)
```

```python
learn = Learner(digit_dls, simple_net, opt_func=SGD, loss_func=digit_recognition_rmse_loss, metrics=batch_accuracy)
```

```python
lr = 0.075
```

```python
learn.fit(150, lr = lr)
```
